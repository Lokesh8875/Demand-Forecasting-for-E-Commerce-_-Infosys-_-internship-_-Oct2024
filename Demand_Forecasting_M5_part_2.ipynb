```python
#columns id and store_id are redundant, do let's drop it
df = df.drop(columns=['id','store_id'])
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
df.describe(include=['object'])
```

```python
filtered_df = df[df['item_id'] == 'HOBBIES_1_001']
filtered_df.head()
```

```python
#dropping dept_id as dept_id = cat_id + sub_category_id
df = df.drop(columns=['dept_id'])
```

```python
df.head()
```

```python
#adding records where item_id, cat_id and state_id are the same
aggregated_df = df.groupby(['item_id', 'cat_id', 'state_id'], as_index=False).sum()
aggregated_df.head()
```

```python
aggregated_df.shape
```

```python
#statewise sales trends
m_columns = [f'm_{i}' for i in range(1, 70)]
state_demand = aggregated_df.groupby('state_id')[m_columns].sum()
```

```python
state_demand.head()
```

```python
months = list(range(1, 70))
state_demand.reset_index(inplace=True)
for state in state_demand['state_id']:
  plt.plot(months, state_demand[state_demand['state_id'] == state].iloc[0, 1:], label=state)
plt.xlabel('Month')
plt.ylabel('Total Demand')
plt.title('State-wise Demand Trends Over Time')
plt.legend(title='State')
plt.show()
```

```python
#category_wise sales trends
cat_demand = aggregated_df.groupby('cat_id')[m_columns].sum()
cat_demand.reset_index(inplace=True)
for cat in cat_demand['cat_id']:
  plt.plot(months, cat_demand[cat_demand['cat_id'] == cat].iloc[0, 1:], label=cat)
plt.xlabel('Month')
plt.ylabel('Total Demand')
plt.title('Category-wise Demand Trends Over Time')
plt.legend(title='Category')
plt.show()
```

```python
#for a particular item
for i in range(3):
  sales_data = aggregated_df.iloc[i, 3:].values
  plt.plot(months, sales_data, marker='o')
  plt.title('Demand Over Time for Items')
  plt.xlabel('Month')
  plt.ylabel('Sales')
  plt.grid()
plt.show()
```

```python
#total sales in terms of categories
category_sales = aggregated_df.groupby('cat_id').sum().iloc[:, 3:].sum(axis=1)
category_sales.plot(kind='bar', title='Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Sales by category for month 1
month_cat_sales = df[['cat_id'] + list(aggregated_df.columns[3:])].groupby('cat_id').sum().iloc[:, 0]  # For m_1
month_cat_sales.plot(kind='pie', autopct='%1.1f%%', title='Market Share for Month 1')
plt.show()
```

```python
#total sales in terms of state
state_sales = aggregated_df.groupby('state_id').sum().iloc[:, 3:].sum(axis=1)
state_sales.plot(kind='bar', title='Total Sales by State')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.show()
```

```python
#Sales by state for month 1
month_state_sales = df[['state_id'] + list(aggregated_df.columns[3:])].groupby('state_id').sum().iloc[:, 0]  # For m_1
month_state_sales.plot(kind='pie', autopct='%1.1f%%', title='Market Share for Month 1')
plt.show()
```

```python
#correlation between sales across months
corr_matrix = aggregated_df[m_columns].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Sales Over Time')
plt.show()
```

```python
df_cat_encoded = pd.get_dummies(df['cat_id'], prefix='cat')
df_with_encoded_cat = pd.concat([df[m_columns], df_cat_encoded], axis=1)
corr_matrix_cat = df_with_encoded_cat.corr()
print(corr_matrix_cat[['cat_FOODS', 'cat_HOUSEHOLD', 'cat_HOBBIES']])
```

```python
df_state_encoded = pd.get_dummies(df['state_id'], prefix='state')
df_with_encoded_state = pd.concat([df[m_columns], df_state_encoded], axis=1)
corr_matrix_state = df_with_encoded_state.corr()
print(corr_matrix_state[['state_CA', 'state_TX', 'state_WI']])
```

```python
grouped_by_cat = df.groupby('cat_id')[m_columns].mean()
corr_by_cat = grouped_by_cat.corr()
print(corr_by_cat)
grouped_by_state = df.groupby('state_id')[m_columns].mean()
corr_by_state = grouped_by_state.corr()
print(corr_by_state)
```
